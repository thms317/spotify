# Artists


#### Load data


```python
import pandas as pd

df = pd.read_csv("./data/playlist_stats_clean.csv")
```

```python
from notebook_functions import show

show(df)
```

#### Count total number of artists


```python
from notebook_functions import count_all_items

total_artist_counts = count_all_items(df, count_what="artist_names")

print(f"Total unique artists: {len(total_artist_counts)}")

total_artist_counts.head(10)
```

```python
from notebook_functions import create_wordcloud_from_series

create_wordcloud_from_series(total_artist_counts[:40], title="TOP ARTISTS")
```

```python
from notebook_functions import top_items_added_by

total_artist_counts_by = top_items_added_by(df, count_what="artist_names")

for name in df["added_by"].unique():
    print(f"\n{name}")
    print("Total added artists:", total_artist_counts_by[name]["total_items"])
    print(f"Top artists:\n{total_artist_counts_by[name]['top_items']}")

    create_wordcloud_from_series(
        total_artist_counts_by[name]["top_items_all"][:20], title=f"TOP ARTISTS\n({name})"
    )
```

#### Count total number of genres


```python
total_genres_counts = count_all_items(df, count_what="artist_genres")

print(f"Total unique genres: {len(total_artist_counts)}")

total_genres_counts.head(10)
```

```python
create_wordcloud_from_series(total_genres_counts[:40], title="TOP GENRES")
```
